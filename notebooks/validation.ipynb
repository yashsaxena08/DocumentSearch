```python
def extract_text(folder_path):
    '''
     This function returns a dictionary which
     contains the file name with text of each
     file in key-value pair.
    '''
    # Path of the folder containing files
    file_text = {}
    docs = glob.glob(folder_path + "/*.pdf")
    files = list(map(lambda x:os.path.splitext(os.path.basename(x))[0],docs))
    for idx,input_file in enumerate(docs):
        parserPdf = parser.from_file(input_file)
        reg = parserPdf['content']
        reg = re.sub('[^A-Za-z0-9/]+', ' ', reg)
        reg = re.sub('\s{2,}', ' ', reg)
        all_chunks = reg.split()
        text = []
        for chunk in range(len(all_chunks)//40):
            text.append(' '.join(all_chunks[chunk*40:(chunk+1)*40]))
        file_text[files[idx]] = text
    return file_text
```
